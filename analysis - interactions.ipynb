```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
%matplotlib inline

plt.rcParams["figure.figsize"] = [10,8]

items = ["100", "500", "1000", "1500"]
source = []
for i in items:
    source.append(pd.read_csv("data/log-n" + i + "-w36-c1.0.csv", index_col=0, header=None).transpose())
```

```python
distance = []

for s in source:
    distance.append(s['distance'].tolist())

a = np.array(list(range(0,len(distance[0]))))
x = a.astype(float)

for d, i  in zip(distance, items):
    plt.plot(x, d, label='n='+i)

plt.legend()
plt.show()
```

```python
unique = []

for s in source:
    unique.append(s['unique'].tolist())

for u, i  in zip(unique, items):
    plt.plot(x, u, label='n='+i)

plt.legend()
plt.show()
```

```python
change = []

for s in source:
    change.append(s['change'].tolist())

for c, i  in zip(change, items):
    plt.plot(x, c, label='n='+i)

plt.legend()
plt.show()
```
